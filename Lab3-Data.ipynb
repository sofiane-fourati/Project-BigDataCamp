```python
file = "DETAILED DATA DESCRIPTION.txt"
with open(file, "r") as f:
        data = f.read()
data = str(data).split('\n')
data = [l for l in data if l != '']
data = data[4:]
data = data[:-95]
```

* __Comparaison des distributions:__

Dans cette partie là, on veut repérer les variables dont la distribution est différente selon la réponse de l'individus : Oui intéressé ou Non. Vu l'absence des condition du test anova on va utiliser le test Kruskal-Wallis dont l'hypothèse nulle est que la médiane de population de tous les groupes est égale. C'est une version non-paramétrique de ANOVA.


```python
for i in yes.columns:
    if (stats.kruskal(yes[i], no[i]).pvalue < 0.000000005) :
        print(i, "\t", stats.kruskal(yes[i], no[i]).pvalue )
```
