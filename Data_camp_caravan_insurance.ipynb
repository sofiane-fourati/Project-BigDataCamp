##  <span style="color:#black">Introduction  </span> 


Sending targeted emails to potential customers of a company is considered a very effective way to market a product or service. On the other hand, an abusive sending can exemplify a disturbance of the customers, who become progressively hostile or disinterested with each reception of this variety of "junk mail".
Indeed, the majority of these letters representing nothing for the customer, end up being wasted, not only because it represents a waste of money for the company which devotes important budgets in the aim of retaining its customers and ensuring a competitive intelligence, but also evokes a need to retrain or revise the company's marketing strategies. However, if the company better understands who its potential customers are, it will be more specific about who to contact, so future budgets will be more effective.

It is with this in mind that the Canadian insurance company AssurExperts Inc. (Insurance and Financial Services Firm) entrusts us with a mission and commits us to define a new advertising strategy based on the analysis of the different behaviors of segments of customers, in order to detect the one or those who appear the most
Our mission consists of these tasks:

- To ensure the profitability of a campaign, our intervention also consists in formulating a meticulous description of current and potential customers, and their degree of appetite to subscribe to a caravan insurance policy. In addition, the definition of a classification threshold will be very beneficial for AssurExperts Inc, in order to better manage the marketing pressure.
- Establish a model that optimizes the number of receptive customers to target the interested active clientele.



## Data description :

The characteristics collected from clients are divided into two families of variables as follows:

- 43 socio-demographic attributes: SD1, ... SD43
- 42 product owner attributes: PO44, ... PO85
- The CLASS attribute represents the response of the clients surveyed. The value YES corresponds to those who wish to have mobile homes.
- The STATUS column indicates whether the observation is for the learning phase or the test phase.

We will be using three different text files:
- AssurancExpertInc.txt which contains the data about the clients
- DetailedDataDescription.txt which decodes the true values of the categorical variables and that we will use for the exploration phase.
- a.txt which contains information on the L0 value will also be used for the exploration phase

##  <span style="color:#black">Business quality metric </span> 

**What do we want to predict?** 
- What we want to predict is whether a person is interested in buying a caravan insurance or not, and building a good prediction of that would help reinforce the company's marketing strategy ( retain it's customers and avoid wasting money )

**What will be the quantitative score? How does the quantitative score reflect the quality or value of the prediction?**

Therefore, we have assumed that it is important for the insurance company to correctly classify the 1's, ie it is more important to know who the customers who will buy the insurance policy are. We also assumed that the cost of attempting to acquire a new client is low, that is, the penalty for incorrectly classifying 1 is not high. We give more weight for the recall than the precision in the evaluation of our models. But we do not want to cause discomfort to the customer list so a compromise between the two metrics is what we are going to consider. Finally, we will also consider precision in evaluating our models because we do not want a model that ranks most observations in a same class. 


    TP: Interested client predicted as interested
    FN: Interested client predicted as non interested ( expensive cost )
    FP: Non interested client predicted as interested ( reasonnable cost )
    TN: Non interested predicted as non interested
    P:  Precision
    R:  Recall
    
    Proportion of potential customers lost = FN / ( TP + FN ) = 1 - R 
    



To emphasize how important recall is, penalizing the proportion term we introduced goes without saying. Additionally we decided to penalize the log loss because to some extent this tells us how reliable the probabilistic approach to the model is.We also took into account the precision factor but with a $\beta$ weight significantly lower than $\alpha$ relatively to the use case.

Weighted_Mixed_Score = ll + $\alpha\times P + \beta\times {Prop}$


Taking this score as a reference it would reflect the quality of the prediction by considering the indicators that are valuable and interesting in our use case

-> It is obvious that lower the metric, better the prediction. low values of this score will reflect the quality of the predictions.

**How do we measure the quality or value of the prediction in the selected business?**

For the Marketing department, to check if our prediction has a real positive impact, we could track the clients' feedback, meaning whether they have been receiving offers that are worth considering or if they were not interested at all in the mail they have received from the company. This tool's efficiency would also be easily measurable when looking at the finances. 



**How does the prediction error convert into cost or decreased KPI?**

If we make wrong predictions, the marketing department would make wrong choices that are quiet expensive, either by  sending mail to completely wrong profiles and spamming people by causing them discomfort or else by losing potentially interested people as the information wouldn't get to them. ### Data formating:

```python
import pandas as pd
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
import warnings
from scipy import stats
from IPython.display import HTML
from IPython.display import display
warnings.filterwarnings('ignore')
import squarify
import colorlover as cl
```

```python
df = pd.read_table("data\\AssurancExpertsInc.txt", sep="\t")
df.head()
```

##### For clarity purposes we are going to replace categorical variables by their true values from the description above.

```python
file = "data\\DETAILED-DATA-DESCRIPTION.txt"
with open(file, "r") as f:
        data = f.read()
data = str(data).split('\n')
data = [l for l in data if l != '']
data = data[4:]
data = data[:-95]
```

```python
for i in range(0,len(data)):
    a = data[i].find(" ")
    b = data[i].find(" ", a + 1)
    data[i] = data[i][b+1 :]
data = data + ['CLASS','STATUS']
data[0] = 'Customer Subtype'
data[1] = 'Number of houses'
data[2] = 'Avg size household'
data[3] = 'Avg age'
data[4] = 'Customer main type'
data[5] = 'Roman catholic'
data[6] = 'Protestant'
df.columns = data
l = pd.read_table("data\\a.txt", sep="\t")
l = l.drop('Unnamed: 0', axis=1)
```

```python
for i in range(0, len(df)):
    for j in range(0, len(l)):
        if l['Value'].iloc[j] == df['Customer Subtype'].iloc[i] : 
            df['Customer Subtype'].iloc[i]  = l['Label'].iloc[j]
```

```python
df['Avg age'].replace(to_replace={1 : '20-30 years' , 2 : '30-40 years', 3 : '40-50 years', 4 : '50-60 years', 
                        5 : '60-70 years', 6 : '70-80 years'}, inplace=True)
```

```python
df['Customer main type'].replace(to_replace={1 : 'Successful hedonists' ,
                                      2 : 'Driven Growers', 
                                      3 : 'Average Family',
                                      4 : 'Career Loners', 
                                      5 : 'Living well',
                                      6 : 'Cruising Seniors',
                                      7 : 'Retired and Religeous',
                                      8 : 'Family with grown ups',
                                      9 : 'Conservative families',
                                      10 : 'Farmers'}, inplace=True)
```

```python
for i in df.columns[5:43]:
    df[i].replace(to_replace={0 : '0%', 1 : '1 - 10%', 2 : '11 - 23%' , 3 : '24 - 36%' , 4 : '37 - 49%' ,
                               5 : '50 - 62%' , 6 : '63 - 75%' , 7 : '76 - 88%' ,8 : '89 - 99%', 9 : '100%' }, inplace=True)
```

```python
for i in df.columns[43:64]:
    df[i].replace(to_replace={0 : '0' ,1 : '1 – 49', 2 : '50 – 99', 3 : '100 – 199', 4 : '200 – 499', 5 : '500 – 999',
                               6 : '1000 – 4999', 7 : '5000 – 9999', 8 : '10.000 - 19.999', 9 : '20.000 - ?' }, inplace=True)
```

```python
trainingSetX, trainingSetY = np.array(df.drop('CLASS',axis = 1)), np.array(df['CLASS'].copy())
```

```python
from sklearn.metrics import log_loss
alpha = 3
beta = 1

results = {}
data = []
for name, clf in zip(names, classifiers):
    clf.fit(trainingSetX, trainingSetY)
    
    scores = {}

    pred = clf.predict(testingSetX)
    pred_proba = clf.predict_proba(testingSetX)
    
    conf_mat = confusion_matrix(testingSetY, pred)
    scores["Confusion matrix"] = conf_mat
    
    logloss =  round(log_loss(testingSetY,pred_proba), 3)
    scores["Log Loss"] = logloss
    
    precision = round(precision_score(testingSetY, pred), 3)
    scores["Precision score"] = precision
    
    Prop_Lost_Cust = 1 - round(recall_score(testingSetY, pred), 3)
    scores["Prop_Lost_Cust"] = Prop_Lost_Cust
     
    scores["Mixed"] = logloss + alpha * Prop_Lost_Cust + beta * (1 - precision)

    results[name] = scores
```

We go to the validation of this model with KFold Cross validation:



```python
from sklearn.metrics import roc_curve, auc

cv = StratifiedKFold(n_splits=15,shuffle=True, random_state = 5)
clf =  LogisticRegression()
tprs = []
aucs = []
fprlist =  []
i = 0
m = 100
data = []
models = []
test_modelX = []
test_modelY = []

mean_fpr = np.linspace(0, 1, 100)
colors = ['#4abd73', '#e5f775', '#f97979', '#d3ffce',"#0000ff",'#008080', '#ff7373', '#ff7373', '#800080',"#ffa500", '#404040', '#000080', '#f94589', '#31698a',"#808080",'#4abd73', '#e5f775', '#f97979', '#d3ffce',"#0000ff"]
for train, test in cv.split(dfX,dfY):
    df_train = df.iloc[train]
    df_minority = df_train[df_train.CLASS==True].copy()
    df_minority_upsampled = resample(df_minority, 
                                 replace=True,    
                                 n_samples=6000,    
                                 random_state=123) 
    dftrain = pd.concat([df_train, df_minority_upsampled])
    df_trainY = df['CLASS'].copy()
    df_trainX = df.drop('CLASS',axis = 1).copy()

    model = clf.fit(np.array(df_trainX), np.array(df_trainY))
    models.append(model)
    
    predictions = model.predict(dfX[test])
    pred_proba = model.predict_proba(dfX[test])
    
    test_modelX.append(dfX[test])
    test_modelY.append(dfY[test])
    # Compute ROC curve and area the curve
    fpr, tpr, thresholds = roc_curve(dfY[test], predictions)
    fprlist.append(fpr)
    tprs.append(interp(mean_fpr,fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    trace3 = go.Scatter(x=fpr, y=tpr,
                        mode='lines', 
                        line=dict(color=colors[i], width=1),
                        name = "ROC FOLD " + str(i)
                       )
    data.append(trace3)
    
    trace2 = go.Scatter(x=[0, 1], y=[0, 1], 
                    mode='lines', 
                    line=dict(color='navy', width=1, dash='dash'),
                    showlegend=False)
    
    logloss =  round(log_loss(dfY[test],pred_proba), 3)
    
    precision = round(precision_score(dfY[test], predictions), 3)
    
    Prop_Lost_Cust = 1 - round(recall_score(dfY[test], predictions), 3)
     
    mixed = logloss + alpha * Prop_Lost_Cust + beta * (1 - precision)
    if (mixed < m):
        m = mixed
        it = i
    i = i+1
data.append(trace2)

print("Fold: ", it)
print("Mixed = ", m)

fig = go.Figure(data=data)
iplot(fig)
```

### Classifier : 

```python
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.base import BaseEstimator
from sklearn.linear_model import LogisticRegression

from sklearn.utils import resample
import pandas as pd
import numpy as np


class Classifier(BaseEstimator):
    
    def __init__(self):
        self.model = LogisticRegression()
    def fit(self, X, y):
        X = pd.DataFrame(X)
        y.reset_index(drop=True, inplace=True)
        df = pd.concat([X, y], axis = 1)
        df_minority = df[df.CLASS==True].copy()
        df_minority_upsampled = resample(df_minority, 
                                 replace=True,    
                                 n_samples=3000,    
                                 random_state=123)
        df = pd.concat([df, df_minority_upsampled])
        y = df['CLASS'].copy()
        X = df.drop('CLASS',axis = 1).copy()
        self.model.fit(np.array(X), np.array(y))

    def predict_proba(self, X):
        proba = self.model.predict_proba(np.array(X))
        return proba
```
